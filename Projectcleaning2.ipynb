```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas_profiling as pp
import sweetviz as sv
import datetime as dt
import streamlit as st
from streamlit_pandas_profiling import st_profile_report
from pandas_profiling import ProfileReport


pd.set_option('display.max_rows', 50)
pd.set_option('display.max_columns', 50)
```

```python
intake = pd.read_csv('./intake_clean.csv')
intake['datetime'] = pd.to_datetime(intake['datetime'], format='%Y-%m-%d %H:%M:%S')
intake['datetime'] = intake['datetime'].apply(lambda x: dt.datetime.strftime(x, '%Y-%d-%m %H:%M:%S'))
intake['datetime'] = pd.to_datetime(intake['datetime'], format='%Y-%d-%m %H:%M:%S')
intake.sort_values(by=['animal_id','datetime'], inplace=True)
intake.reset_index(inplace=True)
intake.drop(columns='index', inplace=True)
intake.head()
```

```python
outcome = pd.read_csv('./outcome_clean.csv')
outcome['datetime'] = pd.to_datetime(outcome['datetime'], format='%Y-%m-%d %H:%M:%S')
outcome['datetime'] = outcome['datetime'].apply(lambda x: dt.datetime.strftime(x, '%Y-%d-%m %H:%M:%S'))
outcome['datetime'] = pd.to_datetime(outcome['datetime'], format='%Y-%d-%m %H:%M:%S')
outcome.sort_values(by=['animal_id','datetime'], inplace=True)
outcome.reset_index(inplace=True)
outcome.drop(columns='index', inplace=True)
outcome.head()
```

```python
merge = pd.merge_asof(intake, outcome, by=['animal_id'], left_index=True, right_index=True,direction='nearest')
merge.rename(columns={'datetime_x': 'datetime_in','datetime_y': 'datetime_out', 'name_x':'name','animal_type_x':'animal_type','breed_x':'breed','color_x':'color','sex_x':'sex'},inplace=True)
merge.drop(merge.filter(regex='_y$').columns.tolist(),axis=1, inplace=True)    
```

```python
merge_times = merge.copy() # creating a copy of my full data, dropping all seperate time columns for now
```

```python
merge.drop(columns=['day_in','month_in','year_in','time_in','date_of_birth','day_out','month_out','year_out','time_out'	] ,inplace=True) # also dropping date of birth, as it is redundant in the age columns
merge.head()
```

```python
merge['days_in_shelter'] = (merge['datetime_out'] - merge['datetime_in'])
merge.head()
```

```python
merge['days_in_shelter']=pd.to_timedelta(merge['days_in_shelter']).dt.total_seconds()
merge['days_in_shelter']=(merge['days_in_shelter']/86400).round(2)
merge.head(10)
```

```python
merge['days_in_shelter']=merge['days_in_shelter'].astype(str)
merge.head()
```

```python
merge['days_in_shelter'].replace('-','', regex=True,inplace=True)
merge.head()
```

```python
merge.dropna(axis=0,inplace=True) #dropping rows with NaN's as those are only around 350 rows, where there was no entry in "intake", so I dropped them as I don't think they are worth the effort of dealing with all those NaN's in regard to the size of the overall data.
merge.shape
```

```python
merge['days_in_shelter']= merge['days_in_shelter'].astype(float)
```
